```python
import pandas as pd
import matplotlib.pyplot as plt

# Totais anuais
totais = {
    "Anos": [2018, 2019, 2020, 2021, 2022],
    "Casamentos": [960, 958, 558, 837, 919],
    "Divrcios": [568, 590, 570, 603, 581],
    "Nascimentos": [2253, 2131, 2102, 2043, 2068],
    "bitos": [2295, 2271, 2439, 2366, 2712],
    "Emigrao": [1656, 1492, 1571, 1312, 1587],
}

# DataFrame
df_totais = pd.DataFrame(totais)
df_totais.set_index("Anos", inplace=True)

# Cores
cores = {
    "Casamentos": "royalblue",
    "Divrcios": "darkorange",
    "Nascimentos": "seagreen",
    "bitos": "firebrick",
    "Emigrao": "purple"
}

# Grfico
ax = df_totais.plot(kind="line", marker="o", figsize=(14, 10), color=[cores[col] for col in df_totais.columns])

# Personalizao
ax.set_title("Demografia Aoriana (20182022)", fontsize=14)
ax.set_ylabel("Quantidade")
ax.legend(title="", loc="upper left")

# Valores
offsets = {
    "Casamentos": 15,
    "Divrcios": -15,
    "Nascimentos": 25,
    "bitos": -25,
    "Emigrao": 20
}

for col in df_totais.columns:
    for x, y in zip(df_totais.index, df_totais[col]):
        ax.annotate(
            str(y),
            (x, y),
            textcoords="offset points",
            xytext=(0, offsets[col]),  #Notas
            ha='center',
            fontsize=10,
            fontweight="bold",
            color=cores[col]
        )

plt.tight_layout()
plt.show()
```
