**Análise sexo e idade por turma**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dados Sexo e Idade
dados_combinados = {
    'Turma': ['7ºA', '7ºB', '7ºFPB3', '8ºA', '8ºB', '8ºFPB3', '9ºA', '9ºB'],

    # Dados de Sexo
    'Feminino': [9, 9, 2, 9, 9, 1, 8, 3],
    'Masculino': [9, 7, 4, 7, 7, 4, 7, 10],
    'Total_Sexo': [18, 16, 6, 16, 16, 5, 15, 13],

    # Dados de Idade (Médias)
    'Media_Idade': [13.1, 12.7, 15.8, 13.6, 13.9, 15.8, 16.1, 14.7],

    # Faixas etárias predominantes
    'Faixa_Etaria': ['13 anos', '12 anos', '16 anos', '13 anos', '14 anos', '15-16 anos', '15 anos', '14 anos']
}

df_combinado = pd.DataFrame(dados_combinados)

# 1. GRÁFICO DUPLO - SEXO E IDADE POR TURMA
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
fig.suptitle('Análise sexo e idade por turma',
             fontsize=16, fontweight='bold')

# Gráfico 1: Distribuição por Sexo
x_pos = np.arange(len(df_combinado['Turma']))
largura = 0.35

bars_f = ax1.bar(x_pos - largura/2, df_combinado['Feminino'], largura,
                label='Feminino', color='#FF69B4', alpha=0.8)
bars_m = ax1.bar(x_pos + largura/2, df_combinado['Masculino'], largura,
                label='Masculino', color='#4169E1', alpha=0.8)

ax1.set_ylabel('Número de Alunos', fontweight='bold', fontsize=12)
ax1.set_title('DISTRIBUIÇÃO POR SEXO', fontsize=14, fontweight='bold')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(df_combinado['Turma'], fontweight='bold')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

#Valores e totais
for i, (f, m, total) in enumerate(zip(df_combinado['Feminino'], df_combinado['Masculino'], df_combinado['Total_Sexo'])):
    ax1.text(i - largura/2, f + 0.1, str(f), ha='center', va='bottom', fontweight='bold', color='#FF1493', fontsize=9)
    ax1.text(i + largura/2, m + 0.1, str(m), ha='center', va='bottom', fontweight='bold', color='#1E90FF', fontsize=9)
    ax1.text(i, total + 0.3, f'Total: {total}', ha='center', va='bottom', fontweight='bold', color='black', fontsize=8)

# Gráfico 2: Média de Idade por Turma
bars_idade = ax2.bar(df_combinado['Turma'], df_combinado['Media_Idade'],
                     color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57', '#FF9FF3', '#54A0FF', '#5F27CD'],
                     alpha=0.8)

ax2.set_ylabel('Média de Idade (anos)', fontweight='bold', fontsize=12)
ax2.set_xlabel('Turmas', fontweight='bold', fontsize=12)
ax2.set_title('MÉDIA DE IDADE POR TURMA', fontsize=14, fontweight='bold')
ax2.grid(axis='y', alpha=0.3)

# Valores de idade
for bar, media, faixa in zip(bars_idade, df_combinado['Media_Idade'], df_combinado['Faixa_Etaria']):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
            f'{media:.1f} anos\n({faixa})', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

```

**Relação: Porporções feminina e masculino**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dados COMBINADOS - Sexo e Idade
dados_combinados = {
    'Turma': ['7ºA', '7ºB', '7ºFPB3', '8ºA', '8ºB', '8ºFPB3', '9ºA', '9ºB'],

    # Dados de Sexo
    'Feminino': [9, 9, 2, 9, 9, 1, 8, 3],
    'Masculino': [9, 7, 4, 7, 7, 4, 7, 10],
    'Total_Sexo': [18, 16, 6, 16, 16, 5, 15, 13],

    # Dados de Idade (Médias)
    'Media_Idade': [13.1, 12.7, 15.8, 13.6, 13.9, 15.8, 16.1, 14.7]
}

df_combinado = pd.DataFrame(dados_combinados)

# Calcular percentuais
df_combinado['%_Feminino'] = (df_combinado['Feminino'] / df_combinado['Total_Sexo'] * 100).round(1)
df_combinado['%_Masculino'] = (df_combinado['Masculino'] / df_combinado['Total_Sexo'] * 100).round(1)

# 1. GRÁFICO DE DISPERSÃO - DUPLA VISÃO
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))
fig.suptitle('RELAÇÃO: IDADE vs PROPORÇÃO POR SEXO POR TURMA',
             fontsize=16, fontweight='bold', y=0.95)

# Gráfico 1: Proporção Feminina vs Idade
scatter_f = ax1.scatter(df_combinado['Media_Idade'], df_combinado['%_Feminino'],
                       s=df_combinado['Total_Sexo']*25,  # Tamanho proporcional ao total
                       c=df_combinado['%_Feminino'], cmap='RdPu', alpha=0.8,
                       edgecolors='black', linewidth=1.5)

# Etiquetas e valores para FEMININO
for i, turma in enumerate(df_combinado['Turma']):
    ax1.annotate(f"{turma}\n{df_combinado['%_Feminino'][i]}%",
                (df_combinado['Media_Idade'][i], df_combinado['%_Feminino'][i]),
                xytext=(8, 8), textcoords='offset points',
                fontweight='bold', fontsize=9,
                bbox=dict(boxstyle="round,pad=0.3", facecolor='#FFB6C1', alpha=0.8),
                ha='left', va='bottom')

ax1.set_xlabel('Média de Idade (anos)', fontweight='bold', fontsize=12)
ax1.set_ylabel('Percentagem Feminino (%)', fontweight='bold', fontsize=12)
ax1.set_title('PROPORÇÃO FEMININA vs IDADE\n(Tamanho = total de alunos)',
             fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=50, color='red', linestyle='--', alpha=0.7, label='Equilíbrio (50%)')
ax1.legend()

# Barra de cores para feminino
cbar_f = plt.colorbar(scatter_f, ax=ax1)
cbar_f.set_label('Percentagem Feminino (%)', fontweight='bold')

# Gráfico 2: Proporção Masculina vs Idade
scatter_m = ax2.scatter(df_combinado['Media_Idade'], df_combinado['%_Masculino'],
                       s=df_combinado['Total_Sexo']*25,  # Mesmo tamanho
                       c=df_combinado['%_Masculino'], cmap='Blues', alpha=0.8,
                       edgecolors='black', linewidth=1.5)

# Etiquetas e valores para MASCULINO
for i, turma in enumerate(df_combinado['Turma']):
    ax2.annotate(f"{turma}\n{df_combinado['%_Masculino'][i]}%",
                (df_combinado['Media_Idade'][i], df_combinado['%_Masculino'][i]),
                xytext=(8, 8), textcoords='offset points',
                fontweight='bold', fontsize=9,
                bbox=dict(boxstyle="round,pad=0.3", facecolor='#87CEEB', alpha=0.8),
                ha='left', va='bottom')

ax2.set_xlabel('Média de Idade (anos)', fontweight='bold', fontsize=12)
ax2.set_ylabel('Percentagem Masculino (%)', fontweight='bold', fontsize=12)
ax2.set_title('PROPORÇÃO MASCULINA vs IDADE\n(Tamanho = total de alunos)',
             fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=50, color='red', linestyle='--', alpha=0.7, label='Equilíbrio (50%)')
ax2.legend()

# Barra de cores para masculino
cbar_m = plt.colorbar(scatter_m, ax=ax2)
cbar_m.set_label('Percentagem Masculino (%)', fontweight='bold')

# Limites dos eixos
for ax in [ax1, ax2]:
    ax.set_xlim(12, 17)
    ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()

# 2. GRÁFICO COMBINADO - VISÃO COMPARATIVA DIRETA
fig, ax = plt.subplots(figsize=(15, 8))

#Plot duplo
scatter_f = ax.scatter(df_combinado['Media_Idade'], df_combinado['%_Feminino'],
                      s=df_combinado['Total_Sexo']*30,
                      c='#FF69B4', alpha=0.7, label='Feminino',
                      edgecolors='darkred', linewidth=2, marker='o')

scatter_m = ax.scatter(df_combinado['Media_Idade'], df_combinado['%_Masculino'],
                      s=df_combinado['Total_Sexo']*30,
                      c='#4169E1', alpha=0.7, label='Masculino',
                      edgecolors='darkblue', linewidth=2, marker='s')

# Linhas de conexão entre pontos da mesma turma
for i, turma in enumerate(df_combinado['Turma']):
    ax.plot([df_combinado['Media_Idade'][i], df_combinado['Media_Idade'][i]],
            [df_combinado['%_Feminino'][i], df_combinado['%_Masculino'][i]],
            'gray', alpha=0.3, linestyle='--')

    # Etiquetas da turma no ponto médio
    mid_point = (df_combinado['%_Feminino'][i] + df_combinado['%_Masculino'][i]) / 2
    ax.annotate(turma,
                (df_combinado['Media_Idade'][i], mid_point),
                xytext=(0, 15), textcoords='offset points',
                fontweight='bold', fontsize=10, ha='center',
                bbox=dict(boxstyle="round,pad=0.2", facecolor='white', alpha=0.8))

ax.set_xlabel('Média de Idade (anos)', fontweight='bold', fontsize=12)
ax.set_ylabel('Percentagem por Sexo (%)', fontweight='bold', fontsize=12)
ax.set_title('RELAÇÃO COMPARATIVA: IDADE vs PROPORÇÃO FEMININA E MASCULINA\nLinhas conectam a mesma turma | Tamanho = total de alunos',
             fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.axhline(y=50, color='red', linestyle='-', alpha=0.5, label='Linha de Equilíbrio (50%)')

#Legenda personalizada
legend1 = ax.legend(loc='upper right')
ax.add_artist(legend1)

# Legenda para tamanhos
for total in [5, 10, 15]:
    ax.scatter([], [], c='gray', alpha=0.5, s=total*30,
              label=f'{total} alunos', edgecolors='black')
ax.legend(scatterpoints=1, frameon=True, labelspacing=1.5,
         title='Tamanho = Total Alunos', loc='lower right')

plt.tight_layout()
plt.show()

# 3. ANÁLISE DE CORRELAÇÃO E TENDÊNCIAS
print("\n" + "="*80)
print("ANÁLISE DE CORRELAÇÃO: IDADE vs DISTRIBUIÇÃO POR SEXO")
print("="*80)

# Correlações
correlacao_idade_feminino = np.corrcoef(df_combinado['Media_Idade'], df_combinado['%_Feminino'])[0,1]
correlacao_idade_masculino = np.corrcoef(df_combinado['Media_Idade'], df_combinado['%_Masculino'])[0,1]

print(f"Correlação Idade vs % Feminino: {correlacao_idade_feminino:.3f}")
print(f"Correlação Idade vs % Masculino: {correlacao_idade_masculino:.3f}")

# Análise de tendências
print("\nTENDÊNCIAS IDENTIFICADAS:")
print("-" * 50)

# Agrupar por faixa etária
df_combinado['Faixa_Etaria'] = pd.cut(df_combinado['Media_Idade'],
                                     bins=[12, 13.5, 15, 17],
                                     labels=['12-13.5 anos', '13.5-15 anos', '15-17 anos'])

for faixa in df_combinado['Faixa_Etaria'].unique():
    turmas_faixa = df_combinado[df_combinado['Faixa_Etaria'] == faixa]
    media_feminino = turmas_faixa['%_Feminino'].mean()
    media_masculino = turmas_faixa['%_Masculino'].mean()

    print(f"\n{faixa}:")
    print(f"  • Proporção média feminina: {media_feminino:.1f}%")
    print(f"  • Proporção média masculina: {media_masculino:.1f}%")
    print(f"  • Turmas: {', '.join(turmas_faixa['Turma'].tolist())}")

# Identificar padrões interessantes
print("\nPADRÕES DESTACADOS:")
print("-" * 30)
equilibradas = df_combinado[(df_combinado['%_Feminino'] >= 45) & (df_combinado['%_Feminino'] <= 55)]
femininas = df_combinado[df_combinado['%_Feminino'] > 55]
masculinas = df_combinado[df_combinado['%_Masculino'] > 55]

print(f"Turmas equilibradas (45%-55%): {', '.join(equilibradas['Turma'].tolist())}")
print(f"Turmas com maioria feminina (>55%): {', '.join(femininas['Turma'].tolist())}")
print(f"Turmas com maioria masculina (>55%): {', '.join(masculinas['Turma'].tolist())}")

# 4. GRÁFICO FINAL RESUMO - BARRAS HORIZONTAIS COMPARATIVAS
fig, ax = plt.subplots(figsize=(14, 8))

# Ordenar por idade
df_ordenado = df_combinado.sort_values('Media_Idade')

y_pos = np.arange(len(df_ordenado))
largura = 0.35

# Barras para feminino e masculino
bars_f = ax.barh(y_pos - largura/2, df_ordenado['%_Feminino'], largura,
                label='Feminino', color='#FF69B4', alpha=0.8)
bars_m = ax.barh(y_pos + largura/2, df_ordenado['%_Masculino'], largura,
                label='Masculino', color='#4169E1', alpha=0.8)

ax.set_yticks(y_pos)
ax.set_yticklabels([f"{turma}\n{idade:.1f} anos" for turma, idade in
                   zip(df_ordenado['Turma'], df_ordenado['Media_Idade'])], fontweight='bold')
ax.set_xlabel('Percentagem (%)', fontweight='bold', fontsize=12)
ax.set_title('DISTRIBUIÇÃO POR SEXO POR TURMA (Ordenado por Idade Média)\nCada turma mostra a idade média',
             fontsize=14, fontweight='bold')
ax.legend()
ax.grid(axis='x', alpha=0.3)
ax.axvline(x=50, color='red', linestyle='--', alpha=0.5)

# Adicionar valores percentuais
for i, (f, m) in enumerate(zip(df_ordenado['%_Feminino'], df_ordenado['%_Masculino'])):
    ax.text(f + 1, i - largura/2, f'{f}%', va='center', fontweight='bold',
           color='#FF1493', fontsize=9)
    ax.text(m + 1, i + largura/2, f'{m}%', va='center', fontweight='bold',
           color='#1E90FF', fontsize=9)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("CONCLUSÕES")
print("="*60)
print("A análise revela que:")
print("• Não há correlação forte entre idade e distribuição por sexo")
print("• As turmas mostram distribuições equilibradas na maioria dos casos")
print("• As turmas FPB3 têm características distintas das regulares")
print("• Cada turma apresenta um perfil único de idade e gênero")
```
