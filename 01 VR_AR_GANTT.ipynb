```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Dados
pesquisadores = [
    {"nome": "Mark Bolas", "publicacoes": "132", "ano_inicio": 1989, "ano_fim": 2015},
    {"nome": "Oliver Bimber", "publicacoes": "175", "ano_inicio": 2003, "ano_fim": 2024},
    {"nome": "Albert Rizzo", "publicacoes": "416", "ano_inicio": 2004, "ano_fim": 2022},
    {"nome": "Olaf Blanke", "publicacoes": "570", "ano_inicio": 1997, "ano_fim": 2025},
    {"nome": "Daniel Thalmann", "publicacoes": "+600", "ano_inicio": 1995, "ano_fim": 2023},
    {"nome": "Mark Billinghurst", "publicacoes": "+785", "ano_inicio": 1993, "ano_fim": 2025},
    {"nome": "Giuseppe Riva", "publicacoes": "1.288", "ano_inicio": 1997, "ano_fim": 2025}
]

# Cores por pesquisador
cores = ["#4CAF50", "#2196F3", "#FFC107", "#E91E63", "#9C27B0", "#00BCD4", "#FF5722"]

# Gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Barras Gantt por pesquisador
for i, p in enumerate(pesquisadores):
    ax.barh(i, p["ano_fim"] - p["ano_inicio"], left=p["ano_inicio"],
            color=cores[i % len(cores)], edgecolor="black")
    ax.text((p["ano_inicio"] + p["ano_fim"]) / 2, i, p["publicacoes"],
            va="center", ha="center", fontsize=9, color="black", fontweight="bold")

# Eixo Y
ax.set_yticks(range(len(pesquisadores)))
ax.set_yticklabels([p["nome"] for p in pesquisadores])

# Eixo X
anos_inicio = list(range(1985, 2026, 10))
ax.xaxis.set_major_locator(ticker.FixedLocator(anos_inicio))
ax.set_xticklabels([f"{ano}-{ano+10}" for ano in anos_inicio], rotation=45)

ax.set_xlim(1985, 2025)

# Rótulos e título
ax.set_xlabel("Período")
ax.set_title("Publicações científicas por pesquisador")

plt.tight_layout()
plt.show()
```
