```python
import matplotlib.pyplot as plt
import numpy as np

# Dados completos de financiamento (em milhes de euros)
empresas = ['Niantic', 'Varjo', 'Ultraleap', 'Mojo Vision', 'Meta (Reality Labs)']
financiamento = [853, 161, 107, 139, 16800]  # Valor estimado para a Meta

# Configurao do grfico
plt.figure(figsize=(12, 8))
bars = plt.bar(empresas, financiamento,
               color=['#2E86AB', '#2E86AB', '#2E86AB', '#2E86AB', '#2E86AB'],
               alpha=0.8, edgecolor='black', linewidth=1)

# Personalizao
plt.title('Investimento Prprio/Recebido', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Milhes de Euros ( M)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Ajustar limites do eixo Y para dar espao aos rtulos
plt.ylim(0, max(financiamento) * 1.15)

# Adicionar valores nas barras com melhor posicionamento
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + (max(financiamento) * 0.02),
             f' {height:,.0f} M'.replace(',', ' '),
             ha='center', va='bottom', fontweight='bold', fontsize=11)

# Adicionar nota explicativa
plt.figtext(0.5, 0.01,
            'Nota: O valor da Meta representa uma estimativa do investimento total em Reality Labs.',
            ha='center', fontsize=10, style='italic')

# Ajustar layout
plt.tight_layout()
plt.show()
```
