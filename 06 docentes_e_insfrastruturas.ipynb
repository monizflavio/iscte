```python
# ========== 1. dados da tabela ==========
labels = ['Pré-Escolar', '1.º Ciclo', 'Port./Línguas', 'Ciências/Mat.',
          'Artes/Ed.Fís.', 'Humanidades/EMC']
sizes  = [20, 39, 30, 24, 19, 15]          # respetivas quantidades
colors = ['#8dd3c7','#ffffb3','#bebada','#fb8072','#80b1d3','#fdb462']

# ========== 2. desenhar o gráfico ==========
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5,5))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
ax.axis('equal')
plt.title('Distribuição dos 147 docentes por área curricular', pad=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

# ---------- dados ----------
labels = ['Dispositivos Móveis', 'PCs Desktop', 'Infraestrutura Rede', 'Servidores']
sizes  = [199, 48, 26, 1]
colors = ['#8dd3c7', '#ffffb3', '#bebada', '#fb8072']
explode = (0, 0, 0.10, 0.15)   # afasta as duas fatias pequenas

# ---------- gráfico ----------
fig, ax = plt.subplots(figsize=(5,5))
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.1f%%',
       colors=colors,
       startangle=90,
       pctdistance=0.85)

ax.axis('equal')
plt.title('Distribuição de equipamentos TIC por categoria')
plt.tight_layout()
plt.show()
```
